```python
def nykaa_tweets(tweets):

  df= pd.DataFrame()

  #df['Column title']= list (map(lambda tweet: tweet.user. ......  ,tweets))

  df['ID']= list (map(lambda tweet: tweet.id,tweets))
  df['TEXT']= list (map(lambda tweet: tweet.text,tweets))
  df['PLACE']= list (map(lambda tweet: tweet.user.location,tweets))
  df['VERIFIED USER']= list (map(lambda tweet: tweet.user.verified,tweets))
  df[' FOLLOWER COUNT']= list (map(lambda tweet: tweet.user.verified,tweets))
  df['FRIENDS COUNT']= list (map(lambda tweet: tweet.user.verified,tweets))
  df['DESCRIPTION']= list (map(lambda tweet: tweet.user.description,tweets))
  df['RETWEET COUNT']= list (map(lambda tweet: tweet.user.lang,tweets))
  #follower counut proportion is missing
  return df
```
